```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
```

```python
df=pd.read_excel("Toronto.xlsx")
```

```python
df.head()
```

```python
df['Borough'].replace('Not assigned', np.nan, inplace=True)
df.dropna(subset=['Borough'], inplace=True)

df.head()
```

```python
df1=df.groupby(["Postal Code","Borough"])["Neighborhood"].apply(". ".join).reset_index()
df1.columns = ['Postal Code', 'Borough', 'Neighborhood']
df1
```

```python
df1['Neighborhood'].replace('Not assigned', "Queen's Park", inplace=True)
df1
```

```python
df1.shape
```

```python
df_geo = pd.read_csv('http://cocl.us/Geospatial_data')
df_geo.columns = ['Postal Code', 'Latitude', 'Longitude']
```

```python
df_pos = pd.merge(df1, df_geo, on=['Postal Code'], how='inner')

df_tor = df_pos[['Borough', 'Neighborhood', 'Postal Code', 'Latitude', 'Longitude']].copy()

df_tor.head()
```

```python
!conda install -c conda-forge geopy --yes
from geopy.geocoders import Nominatim
```

```python
address = 'Toronto, Canada'

geolocator = Nominatim()
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of the City of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
!conda install -c conda-forge folium=0.5.0 --yes
import folium
```

```python
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

for lat, lng, borough, neighborhood in zip(df_tor['Latitude'], df_tor['Longitude'], df_tor['Borough'], df_tor['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=3,
        popup=label,
        color='green',
        fill=True,
        fill_color='#3199cc',
        fill_opacity=0.3,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
